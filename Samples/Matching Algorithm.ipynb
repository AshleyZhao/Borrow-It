### 3.a. Filter out geographical conditions

a and b is the center of the circle, and where the `b_usr` is currently at
r is the active range

```python
def find_geo(x,y,a,b,r):
    # The equation of the range of  ( x − a )^2 + ( y − b )^2 <= r^2
    if ((x - a)**2 + (y - b)**2) <= (r**2):
        return True
    else:
        return False
```

### 3.b. Get timestamp

This function gets the timestamp of the current system and converts the timestamp. In order to use the data, you need the timestamp (total epochs), datetime (a date object), and convert it to string. With the time in string, you can slice it to a usable data.

```python
def get_timestamp():
    x = time.time()
    x = timestamp_to_time(x)
    x = datetime_to_string(x)
    return x[:]
```

This function converts the timestamp given.

```python
def timestamp_to_time(timex):
    return datetime.fromtimestamp(timex)
```

This function converts the datetime to string.

```python
def datetime_to_string(date_time):
    return date_time.strftime("%d-%m-%Y, %H:%M:%S")
```

```python
def update_time(uids, current_time):
    for uid in uids:
        update_db(uid, 'Timestamp', current_time)
```
